# Calculus

## Partial derivative - updation of parameters in gradient descent

```python
import numpy as np

X = np.array([1, 2, 3, 4, 5])
y = np.array([100, 200, 300, 400, 500])

w = 0
b = 0
learning_rate = 0.01
epochs = 100

for epoch in range(epochs):
    predictions = w * X + b

    dw = (1 / len(X)) * np.sum((predictions - y) * X)
    db = (1 / len(X)) * np.sum(predictions - y)

    w -= learning_rate * dw
    b -= learning_rate * db

print("Optimal parameters: \n w =", w, "\n b =", b)
```

## chain rule

```python
import torch
import torch.nn as nn
# neural network architecture
class SimpleNet(nn.Module):
    def __init__(self):
        super(SimpleNet, self).__init__()
        self.hidden = nn.Linear(2, 2)
        self.output = nn.Linear(2, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        a1 = self.sigmoid(self.hidden(x))
        a2 = self.sigmoid(self.output(a1))
        return a2

# set up input, wait, bias

net = SimpleNet()
x = torch.tensor([[0.5, 1.5]], dtype=torch.float32)

# forward pass : Compute Output
output = net(x)
print(f"Neural Network Output: {output.item():.6f}")

#  compute loss and apply chain rule
target = torch.tensor([[1.0]], dtype=torch.float32)

criterion = nn.MSELoss()
loss = criterion(output, target)
print(f"Loss: {loss.item():.6f}")

loss.backward()

#  Access Computed Gradients (Backpropagation)
print("Gradient for hidden weights:\n", net.hidden.weight.grad)
print("Gradient for output weights:\n", net.output.weight.grad)

```
